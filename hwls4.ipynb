```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
t = np.linspace(0,10,51)
t.shape
```

```python
f = np.cos(t)
t.shape
```

```python
plt.plot(t, f, color = 'green')
plt.show()
```

```python
plt.plot(t,f, color = 'green')
plt.title('График f(t).')
plt.xlabel('Значения t.')
plt.ylabel('Значения f.')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

```python
x = np.linspace(-3, 3, 51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
fig.set_size_inches(8,6)
fig.subplots_adjust(wspace=0.5, hspace=0.5) 
ax1.plot(x,y1)
ax1.set_title('График y1.')
ax1.set_xlim([-5, 5])
ax1.set_xlabel('Значения x.')
ax1.set_ylabel('Значения y1.')
ax2.plot(x,y2)
ax2.set_title('График y2.')
ax2.set_xlabel('Значения x.')
ax2.set_ylabel('Значения y2.')
ax3.plot(x,y3)
ax3.set_title('График y3.')
ax3.set_xlabel('Значения x.')
ax3.set_ylabel('Значения y3.')
ax4.plot(x,y4)
ax4.set_title('График y4.')
ax4.set_xlabel('Значения x.')
ax4.set_ylabel('Значения y4.')
```
